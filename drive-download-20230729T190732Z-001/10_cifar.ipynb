# CIFAR-10 Image Data

## Load Libraries and Data

```python
import logging
import io
import requests
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
import matplotlib as mpl
import seaborn as sns

# Restrict tensorflow output to errors
logging.getLogger('tensorflow').setLevel(logging.ERROR)

# Make this notebook's output stable across runs
random_state = 1000
np.random.seed(random_state)
tf.random.set_seed(random_state)

# Plot formatting
%matplotlib inline
sns.set()
mpl.rc('axes', labelsize=14)
mpl.rc('axes', grid=False)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

```python
# Load CIFAR-10 data
((X_train_full, y_train_full),
 (X_test, y_test)) = keras.datasets.cifar10.load_data()

X_valid = X_train_full[:5000] / 255.0
X_train = X_train_full[5000:] / 255.0
y_valid = y_train_full[:5000]
y_train = y_train_full[5000:]
X_test = X_test / 255.0

image_dims = (32, 32, 3)

class_names = [
    'airplane', 'automobile', 'bird',
    'cat', 'deer', 'dog', 'frog',
    'horse', 'ship', 'truck'
]
```

```python
# Output random training instances as images
fig, ax = plt.subplots(1, 6, figsize=(10, 4))
locs = np.random.randint(0, 5000, 6)
for i, a in enumerate(ax):
    a.imshow(X_train[locs[i]])
    a.set_title(class_names[y_train[locs[i]][0]])

plt.show()
```
